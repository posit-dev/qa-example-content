```plaintext
import pandas as pd
import matplotlib.pyplot as plt
import os


# Load the data from CSV
df = pd.read_csv(os.path.join(os.getcwd(), '..', '..', 'data-files', 'spotify_data', 'data.csv'), parse_dates=["release_date"])
```

```plaintext

# List of plotting functions
def create_plots():
    # Plot 1: Histogram of valence
    plt.figure()
    plt.hist(df['valence'], bins=20, color='blue', alpha=0.7)
    plt.title('Valence Distribution')
    plt.xlabel('Valence')
    plt.ylabel('Frequency')
    plt.show()

    # Plot 2: Histogram of year
    plt.figure()
    plt.hist(df['year'], bins=20, color='orange', alpha=0.7)
    plt.title('Year Distribution')
    plt.xlabel('Year')
    plt.ylabel('Frequency')
    plt.show()

    # Plot 3: Scatter plot of danceability vs. energy
    plt.figure()
    plt.scatter(df['danceability'], df['energy'], alpha=0.5, color='green')
    plt.title('Danceability vs Energy')
    plt.xlabel('Danceability')
    plt.ylabel('Energy')
    plt.show()

    # Plot 4: Boxplot of acousticness
    plt.figure()
    plt.boxplot(df['acousticness'])
    plt.title('Acousticness Boxplot')
    plt.ylabel('Acousticness')
    plt.show()

    # Plot 5: Histogram of duration in minutes
    plt.figure()
    plt.hist(df['duration_ms'] / 60000, bins=20, color='purple', alpha=0.7)
    plt.title('Duration (minutes) Distribution')
    plt.xlabel('Duration (minutes)')
    plt.ylabel('Frequency')
    plt.show()

    # Plot 6: Scatter plot of loudness vs. tempo
    plt.figure()
    plt.scatter(df['loudness'], df['tempo'], alpha=0.5, color='red')
    plt.title('Loudness vs Tempo')
    plt.xlabel('Loudness')
    plt.ylabel('Tempo')
    plt.show()

    # Plot 7: Bar plot of mode counts
    plt.figure()
    mode_counts = df['mode'].value_counts()
    plt.bar(mode_counts.index, mode_counts.values, color='cyan')
    plt.title('Mode Counts')
    plt.xlabel('Mode')
    plt.ylabel('Count')
    plt.show()

    # Plot 8: Pie chart of explicit content
    plt.figure()
    explicit_counts = df['explicit'].value_counts()
    plt.pie(explicit_counts, labels=explicit_counts.index, autopct='%1.1f%%', colors=['pink', 'gray'])
    plt.title('Explicit Content')
    plt.show()

    # Plot 9: Line plot of valence over the years
    plt.figure()
    plt.plot(df['year'], df['valence'], alpha=0.7)
    plt.title('Valence Over Years')
    plt.xlabel('Year')
    plt.ylabel('Valence')
    plt.show()

    # Plot 10: Boxplot of tempo
    plt.figure()
    plt.boxplot(df['tempo'])
    plt.title('Tempo Boxplot')
    plt.ylabel('Tempo')
    plt.show()

    # Plot 11: Histogram of instrumentalness
    plt.figure()
    plt.hist(df['instrumentalness'], bins=20, color='lightblue', alpha=0.7)
    plt.title('Instrumentalness Distribution')
    plt.xlabel('Instrumentalness')
    plt.ylabel('Frequency')
    plt.show()

    # Plot 12: Scatter plot of liveness vs speechiness
    plt.figure()
    plt.scatter(df['liveness'], df['speechiness'], alpha=0.5, color='brown')
    plt.title('Liveness vs Speechiness')
    plt.xlabel('Liveness')
    plt.ylabel('Speechiness')
    plt.show()

    # Plot 13: Histogram of loudness
    plt.figure()
    plt.hist(df['loudness'], bins=20, color='yellow', alpha=0.7)
    plt.title('Loudness Distribution')
    plt.xlabel('Loudness')
    plt.ylabel('Frequency')
    plt.show()

    # Plot 14: Line plot of energy over the years
    plt.figure()
    plt.plot(df['year'], df['energy'], alpha=0.7, color='navy')
    plt.title('Energy Over Years')
    plt.xlabel('Year')
    plt.ylabel('Energy')
    plt.show()

    # Plot 15: Bar plot of key counts
    plt.figure()
    key_counts = df['key'].value_counts()
    plt.bar(key_counts.index, key_counts.values, color='olive')
    plt.title('Key Counts')
    plt.xlabel('Key')
    plt.ylabel('Count')
    plt.show()

    # Plot 16: Boxplot of danceability
    plt.figure()
    plt.boxplot(df['danceability'])
    plt.title('Danceability Boxplot')
    plt.ylabel('Danceability')
    plt.show()

    # Plot 17: Histogram of speechiness
    plt.figure()
    plt.hist(df['speechiness'], bins=20, color='teal', alpha=0.7)
    plt.title('Speechiness Distribution')
    plt.xlabel('Speechiness')
    plt.ylabel('Frequency')
    plt.show()

    # Plot 18: Scatter plot of valence vs. acousticness
    plt.figure()
    plt.scatter(df['valence'], df['acousticness'], alpha=0.5, color='maroon')
    plt.title('Valence vs Acousticness')
    plt.xlabel('Valence')
    plt.ylabel('Acousticness')
    plt.show()

    # Plot 19: Histogram of tempo
    plt.figure()
    plt.hist(df['tempo'], bins=20, color='violet', alpha=0.7)
    plt.title('Tempo Distribution')
    plt.xlabel('Tempo')
    plt.ylabel('Frequency')
    plt.show()

    # Plot 20: Bar plot of popularity
    plt.figure()
    plt.bar(df['popularity'].value_counts().index, df['popularity'].value_counts().values, color='lime')
    plt.title('Popularity Distribution')
    plt.xlabel('Popularity')
    plt.ylabel('Frequency')
    plt.show()

# Call the function to generate all plots
create_plots()

```
